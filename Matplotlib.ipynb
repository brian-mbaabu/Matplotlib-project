```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('fuel_econ.csv')

print(fuel_econ.shape)
fuel_econ.head()
```

```python
plt.scatter(data=fuel_econ, x='displ', y='comb');
plt.xlabel('Displacement(l)')
plt.ylabel('Combination');
```

```python
sb.regplot(data=fuel_econ, x='displ', y='comb', x_jitter=0.04,
          scatter_kws ={'alpha' : (1/10)}, fit_reg=False);
```

```python
plt.figure(figsize=[12, 5])

bin_x = np.arange(0, 7+0.3, 0.3)
bin_y = np.arange(12, 58+3, 3)
h2d = plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin=0.5,
          cmap = 'viridis_r', bins = [bin_x, bin_y]);
plt.colorbar()
plt.xlabel('Displacement')
plt.ylabel('Combination in miles per galon');
```

```python
h2d = plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin=0.5,
          cmap = 'viridis_r', bins = [bin_x, bin_y])

counts = h2d[0]
counts.shape
```

```python
plt.figure(figsize = [10, 5])

bin_x = np.arange(0, 7+1, 0.3)
bin_y = np.arange(12, 58+3, 3)
plt.hist2d(data=fuel_econ, x='displ', y='comb', cmin = 0.5, 
           cmap = 'viridis_r', bins = [bin_x, bin_y]);
plt.colorbar()
plt.xlabel('Displacement')
plt.ylabel('Combination miles per gallon');
```

```python
vehicles = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered =True, categories = vehicles)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses);
```

```python
base_color = sb.color_palette()[0]

sb.boxplot(data=fuel_econ, x='VClass', y='comb', color = base_color);

plt.xticks(rotation = 15);
```

```python
fuel_econ.head()
```

```python
fuel_econ['trans_type'] = fuel_econ['trans'].apply(lambda x: x.split()[0])

print(fuel_econ['trans_type'].head())

vehicles = fuel_econ.groupby(['VClass', 'trans_type']).size()
vehicles = vehicles.reset_index(name='count')
vehicles = vehicles.pivot(index ='VClass', columns='trans_type', values='count')
```

```python
plt.subplot(1, 2, 1)
sb.heatmap(vehicles, annot =True, fmt='d');
```

```python
sb.countplot(data=fuel_econ, x='VClass', hue='trans_type');
plt.xticks(rotation=15);
```

```python
base_color = sb.color_palette()[0]
sb.lineplot(data=fuel_econ, x='VClass', y='comb', color=base_color);
plt.xticks(rotation=15);
plt.ylabel("Average Combined Fuel Eff (mpg)");
```
